```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.font_manager as fm

# # this is to update the overall font
# plt.rcParams.update({
#     #"text.usetex": True,
# #     "font.family": "serif",
# #     "font.sans-serif": "Computer Modern",
#     "font.family": "sans-serif",
#     "font.sans-serif": "Poppins",
# })

poppins_font_path = "/Users/robertsalzwedel/Library/Fonts/Poppins-Regular.ttf"  # Change to the correct path if necessary

# Add the font to Matplotlib
fm.fontManager.addfont(poppins_font_path)
plt.rcParams['font.family'] = 'Poppins'  # Use Poppins as the default font


# input data 
data = {'labels': ['A', 'B', 'C', 'D'],
        'values': [20, 35, 25, 20]}
df = pd.DataFrame(data)


# not sure what this is for
#sns.set_style("whitegrid")

# define palette
palette_color = sns.color_palette(['#d2b48c', '#deb887', '#bc8f8f', '#f4a460', '#cd853f'])

# figure definition
fig, ax = plt.subplots(figsize=(7,6))
ax.pie(df['values'], labels=None, autopct='%1.0f%%',colors=palette_color,wedgeprops={'edgecolor': '#f4ecd8'})

# choose background color
fig.patch.set_facecolor('#f4ecd8')  # Light tan for the figure background
ax.set_facecolor('#f4ecd8')  # Light tan for the axes background

# legend
legend = ax.legend(df['labels'],loc="center left", bbox_to_anchor=(1, 0, 0.5, 1), frameon=False,fontsize = 12)

# Add title
plt.title('Climate Pie Chart', fontsize=18, color='darkred')

# Add a footnote-like text to the bottom left corner
ax.text(0.01, 0.01, 'This is where the References are supposed to go.', 
        fontsize=10, 
        color='gray', 
        ha='left', 
        va='bottom', 
        transform=ax.transAxes)

plt.show()
plt.tight_layout()
fig.savefig("Test.pdf")

```
